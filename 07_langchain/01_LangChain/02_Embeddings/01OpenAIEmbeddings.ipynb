### OpenAI Embeddings

```python
from langchain_openai import OpenAIEmbeddings

embedding_model = OpenAIEmbeddings(model="text-embedding-3-large")
embedding_model
```

```python
text = "Hello World. This is a tutorial for the OpenAI embeddings"
query_result = embedding_model.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
## We can also set our own dimesion of the embedding
embedding_model_1024 = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)

query_result = embedding_model_1024.embed_query(text)
query_result

```

```python
len(query_result)
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

loader = TextLoader('speech.txt')
docs = loader.load()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_documents = text_splitter.split_documents(docs)
final_documents
```

```python
## Convert this Document into the vectors using the OpenAI Embeddings and store it into the Vector StoreDB (ChromaDB)
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_documents, embedding_model_1024)
db
```

```python
## We are trying to match this query with the vectors in the vector db

query = "It will be all the easier for us to conduct"
retrieved_results = db.similarity_search(query)
print(retrieved_results)
```
