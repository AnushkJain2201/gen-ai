#### Sentence Transformers on Huggingface
Sentence Transformers are models designed to convert text (sentences, paragraphs, documents) into numerical vectors called embeddings that capture the semantic meaning of the text.

```python
from langchain_huggingface import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text = "this is a test documents"
query_result = embedding_model.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
doc_result = embedding_model.embed_documents([text, "this is not a text document"])
doc_result
```
