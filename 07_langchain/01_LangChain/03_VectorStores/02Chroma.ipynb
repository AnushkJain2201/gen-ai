### Chroma
Chroma is AI Native open source vector database focused on developer productivity and happiness. Chroma is licensed under Apache 2.0

```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter

loader = TextLoader("speech.txt")
docs = loader.load()

text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 0)
splits = text_splitter.split_documents(docs)
splits
```

```python
embeddings = OllamaEmbeddings(model="gemma2:2b")
db = Chroma.from_documents(splits, embedding=embeddings)
db
```

```python
## querying
query = "How does the speaker describe the desired outcome of the war?"
docs = db.similarity_search(query)
docs[0].page_content
```

```python
## Save to the disk and load
db = Chroma.from_documents(splits, embedding=embeddings, persist_directory="./chroma_db")
```

```python
db2 = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)
query = "How does the speaker describe the desired outcome of the war?"
docs = db.similarity_search(query)
docs[0].page_content
```

```python
docs = db.similarity_search_with_score(query)
docs
```

```python
### Retriever option
retriever = db.as_retriever()
retriever.invoke(query)[0].page_content
```
