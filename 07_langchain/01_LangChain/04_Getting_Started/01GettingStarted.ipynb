```python
import os
from dotenv import load_dotenv

load_dotenv(override=True)

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")

## For LangSmith Tracking
os.environ["LANGSMITH_TRACING"] = "true"
os.environ["LANGSMITH_API_KEY"] = os.getenv("LANGSMITH_API_KEY") or os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGSMITH_PROJECT"] = os.getenv("LANGSMITH_PROJECT") or os.getenv("LANGCHAIN_PROJECT")
 
# Usually optional for hosted LangSmith, but safe to set
os.environ["LANGSMITH_ENDPOINT"] = os.getenv("LANGSMITH_ENDPOINT")
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o")
llm
```

```python
## Input & get response from LLM
result = llm.invoke("What is Gen AI")
```

```python
result.content
```

```python
## Chatprompt Template, we can tell our LLM model to behave in a particular way
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are an expert AI Engineer. Provide me answers based on the question"),
        ("user", "{input}")
    ]
)

prompt
```

```python
## here we are creating chain, so basically based on the below code, whatever input we provide, that will first go through our prompt and then passed to the llm model
chain = prompt|llm

response = chain.invoke({
    'input', "Can you tell me about LangSmith?"
})

response
```

```python
## We are getting the response as the AIMessage object, but we can also parse that response in our own way
## We can do that with the help of StrOutputparser
from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()

chain = prompt|llm|output_parser
response = chain.invoke({
    'input', "Can you tell me about LangSmith?"
})

response
 
```
