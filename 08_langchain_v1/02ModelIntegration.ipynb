### Streaming && Batch

#### Streaming

Most models can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experience, particularly for longer responses. Calling stream() returns an iterator that yields output chunks as they are produced. You can use a loop to process each chunk in real-time. 

```python
model.stream("Write a 200 words paragraph on Artificial Intelligence")
```

```python
for chunk in model.stream("Write a 200 words paragraph on Artificial Intelligence"):
    print(chunk.text)
```

#### BATCH

Batching a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done parallely.

```python
responses = model.batch([
    "Why do parrots have colourful feathers?",
    "How do airplane fly?",
    "What is quantum computing?"
])

for response in responses:
    print(response)
```

```python
responses = model.batch([
    "Why do parrots have colourful feathers?",
    "How do airplane fly?",
    "What is quantum computing?"
],
config = {
    "max_concurrency": 5, # Limit to 5 parallel calls
})

for response in responses:
    print(response)
```
